## 棒グラフ
[wiki](https://ja.wikipedia.org/wiki/%E6%A3%92%E3%82%B0%E3%83%A9%E3%83%95)参考<br>
二つ以上の値を比較するのに使う

```python
# wiki記載のデータ作成
array = [['EUL',39,49],\
         ['PES',200,210],\
         ['EFA',42,56],\
         ['EDD',15,19],\
         ['ELDR',67,60],\
         ['EPP',276,272],\
         ['UEN',27,36],\
         ['etc',66,29]]
names = ['Political Party','Y2004','Y1999']
df = pd.DataFrame(array,columns=names)
df = df.set_index('Political Party')
df
```

<hr>
棒グラフ

```python
fig = plt.figure(figsize=(6.4, 4.8), dpi=100,facecolor='g', linewidth=0)
fig.suptitle('Single Bar')

ax1 = fig.add_subplot(1,1,1,fc='gray')
ax1.bar(df.index,df['Y2004'],color='yellow',edgecolor='black',alpha=0.5,hatch='x')
```


<hr>棒グラフ2

```python
fig = plt.figure(figsize=(6.4, 4.8), dpi=100,facecolor='g', linewidth=0)
fig.suptitle('Multi Bars')
ax1 = fig.add_subplot(1,1,1,fc='gray',xlabel=df.index.name,ylabel='count')

x = np.arange(len(df))
w=0.3
ss = ['EUL','PES','EFA','EDD','ELD','EPP','UEN','etc']

ax1.set_xticks(x+w/2)
ax1.set_xticklabels(ss,rotation=60)
ax1.bar(x  ,df['Y1999'],width=0.3,color='blue',edgecolor='black',alpha=0.5,hatch='*')
ax1.bar(x+w,df['Y2004'],width=0.3,color='yellow',edgecolor='black',alpha=0.5,hatch='x')

```

<hr>
積上げ棒グラフ

```python
fig = plt.figure(figsize=(6.4, 4.8), dpi=100,facecolor='g', linewidth=0)
fig.suptitle('Stacked Bar Chart')
ax1 = fig.add_subplot(1,1,1,fc='gray',xlabel=df.index.name,ylabel='count')

x = np.arange(len(df))
ss = ['EUL','PES','EFA','EDD','ELD','EPP','UEN','etc']

ax1.set_xticks(x)
ax1.set_xticklabels(ss,rotation=60)
ax1.bar(x,df['Y1999'],                   color='blue',edgecolor='black',alpha=0.5,hatch='*')
ax1.bar(x,df['Y2004'],bottom=df['Y1999'],color='yellow',edgecolor='black',alpha=0.5,hatch='x')
```

<hr>
積上100%棒グラフ

```python
dfper=df.copy()
dfper["Total"]=dfper["Y2004"]+dfper["Y1999"]
print(dfper)
dfper=dfper.drop("Total", axis=1).divide(dfper["Total"], axis=0)


fig = plt.figure(figsize=(6.4, 4.8), dpi=100,facecolor='g', linewidth=0)
fig.suptitle('Stacked Bar Chart')
ax1 = fig.add_subplot(1,1,1,fc='gray',xlabel=dfper.index.name,ylabel='count')

x = np.arange(len(dfper))
ss = ['EUL','PES','EFA','EDD','ELD','EPP','UEN','etc']

ax1.set_xticks(x)
ax1.set_xticklabels(ss,rotation=60)
ax1.bar(x,dfper['Y1999'],                   color='blue',edgecolor='black',alpha=0.5,hatch='*')
ax1.bar(x,dfper['Y2004'],bottom=dfper['Y1999'],color='yellow',edgecolor='black',alpha=0.5,hatch='x')
```
